# Edge Detection - Canny Edge Detector

It follows a multi step process:
    * Blur the image to avoid noise.
    * Calculating the gradients in X & Y direction.
    * Non-Maxima suppression. (ie., in the neighbourhood, ignore the gradients which are less than the maximum)
    * Hystersis thresholding (ie., obtain a new binary image, where the background is black except for the pixels where gradient is high).

## What is gradient?

The directional change in the intensity value is called the gradient.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
img = cv2.imread('./Images/capsicum.jpg')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
# convert to gray image
gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# workaroud to get the gray image. save and load back.
cv2.imwrite('./Images/gray_capsicum.jpg', gray_img)
```

```python
#?cv2.imwrite
gray_img = cv2.imread('./Images/gray_capsicum.jpg')
```

```python
plt.imshow(gray_img)
plt.title('Gray')
```

```python
# blur the image
blur_img = cv2.GaussianBlur(gray_img, (5,5), 0)
```

```python
plt.imshow(blur_img)
plt.title('Blurred')
```

```python
?cv2.Canny
```

```python
# imge, threshold1 threshold2 for hysteresis procedure
canny = cv2.Canny(blur_img, 30, 150)

# work around.
cv2.imwrite('./Images/canny_capsicum.jpg', canny)
canny = cv2.imread('./Images/canny_capsicum.jpg')
```

```python
plt.imshow(canny)
```
