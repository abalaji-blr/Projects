# Splitting and Merging channels

**Objective:** For a given image, split the channels and show. Also, merge and display them as well.

## Read the Image

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
print('Numpy : ', np.__version__)
print('CV    : ', cv2.__version__)
```

```python
img = cv2.imread('./Images/capsicum.jpg')

# BGR to RGB format
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
img.shape
```

```python
type(img)
```

```python
plt.imshow(img)
plt.show()
```

## Split the image

```python
# note that the image is already converted in RGB format
[R, G, B] = cv2.split(img)
```

```python
type(R)
```

```python
R.shape
```

```python
G.shape
```

```python
?plt.subplot
```

## Show image using subplot()

```python
# show the different channels

# row, column, index
plt.subplot(1,3,1)
plt.imshow(R)
plt.title('Red')

# second fig
plt.subplot(1,3,2)
plt.imshow(G)
plt.title('Green')

# third figure
plt.subplot(1,3,3)
plt.imshow(B)
plt.title('Blue')

plt.tight_layout()
plt.show()
```

```python
?plt.imshow
```

## Show Image using subplots()

```python
f, axes = plt.subplots(1,3, figsize=(12,5))
#axes[0].imshow(cv2.merge[R,0,0])
axes[0].imshow(R)
axes[1].imshow(G)
axes[2].imshow(B)
f.suptitle('Splitting of channels')
plt.show()
```

```python
zeros = np.zeros(R.shape, dtype='uint8')
```

```python
zeros.shape
```

```python
img.shape
```

```python
f, axes = plt.subplots(1,3, figsize=(12,5))
axes[0].imshow(cv2.merge([R,zeros,zeros]))
axes[0].title.set_text('Red')

axes[1].imshow(cv2.merge([zeros,G, zeros]))
axes[1].title.set_text('Green')

axes[2].imshow(cv2.merge([zeros, zeros,B]))
axes[2].title.set_text('Blue')
plt.show()
```
