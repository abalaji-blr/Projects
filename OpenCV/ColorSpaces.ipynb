# Color Spaces 

**Objective:** Explore the different color spaces in OpenCV.

## Read the Image

```python
import cv2
import numpy as np

import matplotlib.pyplot as plt

%matplotlib inline
```

```python
cv2.__version__
```

```python
# opencv reads the image in BGR format
bgr_img = cv2.imread('./Images/capsicum.jpg')
```

```python
type(bgr_img)
```

```python
bgr_img.shape
```

```python
?cv2.cvtColor
```

```python
gray_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2GRAY)
```

```python
gray_img.shape
```

```python
## there seems to be some issue with BGR2GRAY conversion while displaying in matplotlib after conversion.
## workaround is to save to disk and load back.
```

```python
cv2.imwrite('./Images/capsicum_gray.jpg', gray_img)
```

```python
gray_img = cv2.imread('./Images/capsicum_gray.jpg')
```

```python
# hue-saturation-value
hsv_img  = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2HSV)

# luma, A , B
lab_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2LAB)
```

```python
#rgb color
rgb_img = cv2.cvtColor(bgr_img, cv2.COLOR_BGR2RGB)
```

```python
# display them

fig, axes = plt.subplots(2,2, figsize=(10,6))
axes[0,0].imshow(rgb_img)
axes[0,0].title.set_text('RGB')

axes[0,1].imshow(gray_img)
axes[0,1].title.set_text('Gray')

axes[1,0].imshow(hsv_img)
axes[1,0].title.set_text('HSV')

axes[1,1].imshow(lab_img)
axes[1,1].title.set_text('LAB')

plt.tight_layout()
plt.show()
```
