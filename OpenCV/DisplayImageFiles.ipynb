## Load and Display images

```python
import cv2
```

```python
from os import listdir
```

```python
listdir('.')
```

```python
def get_image_files(dir):
    result = []
    for f in listdir(dir):
        if f.endswith('.' + 'jpg'):
            result.append(f)
    return(result)
```

```python
image_files = get_image_files('.')
```

```python
image_files
```

```python
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
?cv2.imread
```

```python
for f in image_files:
    img = cv2.imread(f)
    
    ## BGR format
    #plt.imshow(img)
    #plt.show()
    
    # BGR to RGB format
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
    
```
