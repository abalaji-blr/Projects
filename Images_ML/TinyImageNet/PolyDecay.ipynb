## Find the LR value

```python
maxEpochs=150
```

```python
def poly_decay_ep125(base_lr_val, epoch):
  #maxEpochs = 150
  base_lr   = base_lr_val
  power     = 1.0
  
  # compute new lr
  new_lr = base_lr * ( 1 - (epoch / float(maxEpochs))) ** power
  
  #print(new_lr)
  # new learning rate
  return new_lr
```

```python
for epoch in range(2,maxEpochs):
    new_lr = poly_decay_ep125(1e-1,epoch)
    print('Epoch: %d LR_value: %f' % (epoch, new_lr))
```

```python
for epoch in range(2, 50):
    new_lr = poly_decay_ep125(0.0013,epoch)
    print('Epoch: %d LR_value: %f' % (epoch, new_lr))
```

```python
for epoch in range(2, 50):
    new_lr = poly_decay_ep125(0.00125,epoch)
    print('Epoch: %d LR_value: %f' % (epoch, new_lr))
```
